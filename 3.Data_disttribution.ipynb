```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
data_distribution = np.random.normal(loc=50, scale=10, size=1000)

plt.hist(data_distribution, bins=30, edgecolor='black')
plt.title("Histogram - Data Distribution")
plt.xlabel('values')
plt.ylabel("Frequency")
plt.show()
```
